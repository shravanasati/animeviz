```python
import pandas as pd

df = pd.read_csv("./anime_data.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=["explicit_genres", "nsfw", "start_season"])
df
```

```python
df["status"].value_counts()
```

```python
textcols = [
    "synopsis",
    "genres",
    "themes",
    "demographics",
    "studios",
    "related_anime",
    "alt_title_en",
    "alt_title_jp",
]
for col in textcols:
    df[col] = df[col].fillna("")
```

```python
df.isnull().sum()
```

```python
df["mean"].plot(kind="hist")
```

Imputing the `mean` field with mean due to absence of outliers.

```python
df["mean"] = df["mean"].fillna(df["mean"].mean())
```

```python
df["rank"] = df["rank"].fillna(99999)
```

```python
unknown_cols = ["media_type", "rating"]
for col in unknown_cols:
    df[col] = df[col].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
df["num_scoring_users"] = df["num_scoring_users"].fillna(0)
df["num_episodes"] = df["num_episodes"].fillna(-1)
df["start_date"] = df["start_date"].fillna("")
df["end_date"] = df["end_date"].fillna("")
```

```python
df.isnull().sum().sum()
```

```python
df["synopsis"] = df["synopsis"].apply(lambda x: x.replace("[source]", "").replace("[Written by MAL Rewrite]", "").strip())
```

```python
df.to_csv("anime_data_cleaned.csv", index=False, encoding="utf-8")
```
